```python
mi_nombre= 'Carlos'
mi_apellido= 'Sanchez'
```

```python
registro_sucio = " ##[ID-9875]__Monto:S/.4500__Estado:PENDIENTE##"
```

LIMPIEZA


Paso A: Quita los espacios en blanco y los caracteres ## que están en los
extremos (bordes izquierdo y derecho) del texto usando

```python
registro_sucio = registro_sucio.strip(' #')
registro_sucio
```

Paso B: Convierte todo el texto resultante a minúsculas usando .lower().

```python
registro_sucio = registro_sucio.lower()
```

Paso C: Usa .replace() para cambiar la palabra "pendiente" por la
palabra "procesado".

```python
registro_sucio = registro_sucio.replace('pendiente','procesado')
```

Paso D: Divide el texto limpio en partes usando como separador los guiones
bajos dobles __ mediante el método .split().

```python
registro_limpio = registro_sucio.split('__')
```

Paso E: De la lista resultante del split, extrae la parte que corresponde al monto
e identifica en qué posición (índice) se encuentra la primera letra "m" de la
palabra "monto" usando .index().

```python
registro_limpio
monto= registro_limpio[1]
ind_m = monto.index('m')
```

**Reporte Final**: Usa la función print() junto con una f-string para mostrar en la
pantalla el siguiente mensaje exacto (reemplazando los campos con tus variables y los
datos extraídos):

```python
print(f"""Reporte generado por {mi_nombre} {mi_apellido}.
{registro_limpio}.
La 'm' de monto se encontró en la posición: {ind_m}.""")
```
